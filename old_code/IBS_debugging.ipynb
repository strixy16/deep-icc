```python
import hdfs_config as args
from hdfs_data_loading import *
import hdfs_train
from sksurv.metrics import integrated_brier_score
```

```python
train_info_path = os.path.join(args.DATA_DIR, args.TRAIN_LABEL_FILE)
test_info_path = os.path.join(args.DATA_DIR, args.TEST_LABEL_FILE)

train_info = pd.read_csv(train_info_path)
train_time = np.asarray(train_info['HDFS_Time'])
train_event = np.asarray(train_info['HDFS_Code'])
```

```python
test_predictions_path = '/Data/Output/HDFSModel2/2022_02_02_1706/test_predictions.csv'

pd_test_predictions = pd.read_csv(test_predictions_path)
pd_test_predictions.head()

test_predictions = np.asarray(pd_test_predictions['Prediction'])
test_time = np.asarray(pd_test_predictions['Time'])
test_event = np.asarray(pd_test_predictions['Event'])
```

```python
train_time.shape
```

```python
test_predictions = test_predictions

dt = np.dtype([('event', bool), ('time', float)])
train_surv = np.zeros(train_time.shape[0], dt)
train_surv['event'] = train_event
train_surv['time'] = train_time

test_surv = np.zeros(test_time.shape[0], dt)
test_surv['event'] = test_event
test_surv['time'] = test_time
```

```python
times = np.arange(5, max(test_time))
```

```python
test_predictions
```

```python
score = integrated_brier_score(train_surv, test_surv, test_predictions, times)
```
