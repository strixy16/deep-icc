```python
from lifelines import KaplanMeierFitter, CoxPHFitter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
selected_filename = "../../Data/top8_TextureFeatures.xlsx"

selected_features = pd.read_excel(selected_filename)
```

```python
selected_features.head()
```

```python
feats_4_cph = selected_features.drop(columns=['ScoutID', 'LBP25'])
cph = CoxPHFitter()
cph.fit(feats_4_cph, duration_col='RFS', event_col='RFS_Code')
cph.print_summary(model="Selected ICC Texture Features", decimals=3)
cph.check_assumptions(feats_4_cph, p_value_threshold=0.05, show_plots=True)
```

```python
train_file = "../../Data/train_pats.xlsx"
train_pats = pd.read_excel(train_file)
train_pats.head
```

```python
train_textfeats = pd.merge(train_pats, selected_features, how='inner', on=['ScoutID', 'ScoutID'])
train_textfeats.head
```

```python
feats_4_cph = train_textfeats.drop(columns=['ScoutID', 'LBP25'])
cph = CoxPHFitter()
cph.fit(feats_4_cph, duration_col='RFS', event_col='RFS_Code')
cph.print_summary(model="Selected ICC Texture Features", decimals=3)
cph.check_assumptions(feats_4_cph, p_value_threshold=0.05, show_plots=True)
```
