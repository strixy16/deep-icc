```python
import pandas as pd
import numpy as np

rfs_data = pd.read_excel('../Data/RFS_Scout.xlsx')

rfs_median = np.median(rfs_data.RFS)
rfs_mean = np.mean(rfs_data.RFS)

print("Median: ", rfs_median)
print("Mean: ", rfs_mean)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

data = ax.scatter(rfs_data.ScoutID, rfs_data.RFS, label='RFS')

y_median = [np.median(rfs_data.RFS)]*len(rfs_data.ScoutID)
y_mean = [np.mean(rfs_data.RFS)]*len(rfs_data.ScoutID)

median_line = ax.plot(rfs_data.ScoutID, y_median, label='Median RFS', linestyle='--', color='red')
mean_line = ax.plot(rfs_data.ScoutID, y_mean, label='Mean RFS', linestyle='--', color='green')

legend = ax.legend(loc='upper right')
```
